```python
import tensorflow as tf
import keras.utils.vis_utils as kerutils
from IPython.display import Image
%matplotlib inline
    
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(500, activation=tf.nn.relu),  #100
    tf.keras.layers.Dense(300, activation=tf.nn.relu),  #100  
    tf.keras.layers.Dense(100, activation=tf.nn.relu),  #100  
    tf.keras.layers.Dense(10, activation=tf.nn.softmax) 
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print('plotting model')
kerutils.plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
display(Image(filename='model.png'))
#Image(filename='model.png') 
print('end plotting model')
```
